    Naive Bayes
</h1>
The Naive Bayes algorithm is a classification algorithm that uses Bayes' theorem to make predictions. 
The algorithm is called "naive" because it doesn't account for dependencies between features.

```python
import numpy as np
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.calibration import CalibratedClassifierCV

import warnings
warnings.filterwarnings("ignore")
```

<h1 style="font-size:30px;font-family:Calibri">
    Training a Classifier for Continuous
 Features
</h1>

```python
iris = datasets.load_iris()
features = iris.data
target = iris.target

classifier = GaussianNB()
model = classifier.fit(features, target)
```

```python
new_obs = [[4, 4, 4, 0.4]]
model.predict(new_obs)
```

<h1 style="font-size:30px;font-family:Calibri">
    Training a Classifier for Discrete and
 Count Features
</h1>

```python
text_data = np.array([
    'I love Brazil. Brazil!',
    'Brazil is Best',
    'Germany beats both'
])

count = CountVectorizer()
bag_of_words = count.fit_transform(text_data)
```

```python
features = bag_of_words.toarray()
target = np.array([0, 0, 1])
classifier = MultinomialNB(class_prior = [0.25, 0.5])
model = classifier.fit(features, target)
```

```python
new_obs = [[0, 0, 0, 1, 0, 1, 0]]
model.predict(new_obs)
```

<h1 style="font-size:30px;font-family:Calibri">
    Calibrating Predicted Probabilities
</h1>

```python
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
classifier = GaussianNB()
classifier_sigmoid = CalibratedClassifierCV(classifier, cv = 2, method = "sigmoid")
classifier_sigmoid.fit(features, target)
new_obs = [[2.6, 2.6, 2.6, 0.4]]
classifier_sigmoid.predict_proba(new_obs)
```

```python
classifier.fit(features, target).predict_proba(new_obs)
```
